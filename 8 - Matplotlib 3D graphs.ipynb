## Matplotlib 3D graphs
- Create 3D graphs with matplotlib (lines, scatter, bars)



```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import style
import random
import numpy as np 

style.use('ggplot')
```

```python
# simple 3D line 
plt.clf()
fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(111, projection='3d')

n_points = 10

x = range(n_points)
y = [random.random()*i for i in range(n_points)]
z = [random.random()*i for i in range(n_points)]

ax1.plot_wireframe(x,y,z)

ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')

plt.show()
```

```python
# 3D scatter plot

plt.clf()

fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(111, projection='3d')

x2 = range(10)
y2 = [random.random() for i in range(n_points)]
z2 = [random.random() for i in range(n_points)]

ax1.plot_wireframe(x,y,z, color='r')
ax1.scatter(x, y, z, c='c', marker='+')
ax1.scatter(x2, y2, z2, c ='k', marker='o')

ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')

plt.show()
```

```python
# 3D bar plot 
style.use('fivethirtyeight')

plt.clf()

fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(111, projection='3d')

n_bars = 10

# starting points
x3 = range(n_bars)
y3 = range(n_bars) # having some depth on y
z3 = np.zeros(n_bars) # if these are not all zero, bar would be floating

# depths
dx = np.ones(n_bars)*0.8 # some space between bars
dy = np.ones(n_bars) * 0.8 # one depth on y
dz = [random.random() for i in range(n_bars)] # these are our bars' heights

ax1.bar3d(x3, y3, z3, dx, dy, dz, color='c')


ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')

plt.show()
```

```python
# Plot a simple wireframe
plt.clf() 

fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(111, projection='3d')

x, y, z = axes3d.get_test_data()

ax1.plot_wireframe(x,y,z, rstride = 2, cstride = 2, color='c')

ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')

plt.show()
```
