```python
df_demographics = pd.read_csv('data/demographics.csv', delimiter=';', quotechar='"', index_col=0)
df_demographics.describe()
```

```python
# Renaming columns for direct access
df_demographics.rename(columns={'GDP per capita':'GDP_per_capita', 'Life expectancy': 'Life_expectancy'}, inplace=True)
df_demographics.GDP_per_capita
```

```python
# Filtering two lines (GDP per capita, Life expectancy and population in 2013) by index (Country)
df_filtered = df_demographics.loc[['Brazil', 'Albania'], ['GDP_per_capita', 'Life_expectancy', '2013']]
df_filtered
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

# Filtering BRICS countries
df_filtered = df_demographics.loc[['Brazil', 'Russia', 'India', 'China', 'South Africa']]
# Turning years into indexes to plot graph :)
df_filtered = df_filtered.transpose()
# Removing columns of GDP, Life expetancy and old index
df_filtered = df_filtered.drop(df_filtered.index[:3])/1000000

# Set style and create two subplots
style.use('ggplot')
ax1 = plt.subplot2grid((2,1), (0,0))
ax2 = plt.subplot2grid((2,1), (1,0), sharex=ax1)

# We're going to plot India and China separately as they have really higher populations
df_filtered[['China']].plot(ax=ax1, color='r')
df_filtered[['India']].plot(ax=ax1, color='y')
df_filtered[['Brazil']].plot(ax=ax2, color='g')
df_filtered[['Russia']].plot(ax=ax2, color='k')
df_filtered[['South Africa']].plot(ax=ax2, color='b')
plt.ylabel('Population (in millions)')
plt.xlabel('Years')
plt.title('BRICS population ('+df_filtered.index[0]+'-'+df_filtered.index[-1]+')', y=2.2)
plt.legend(loc='best')
plt.show()
```

```python
# Concatenating DataFrames 

df1 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2001, 2002, 2003, 2004])

df2 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2005, 2006, 2007, 2008])

df3 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'Low_tier_HPI':[50, 52, 50, 53]},
                   index = [2001, 2002, 2003, 2004])

# if columns are the same, no problems:
df_concat = pd.concat([df1,df2])
df_concat
```

```python
# What if we have different columns?
# pd.concat has the same effect of append method 
df_concat = df1.append(df3)
df_concat
```

```python
# Creating new row (pandas Series) and appending to another DataFrame
s = pd.Series([80,2,50], index=['HPI','Int_rate','US_GDP_Thousands'])
df_concat = df1.append(s, ignore_index=True)
df_concat
```

```python
# Concatenating DataFrames 

df1 = pd.DataFrame({'Year':[1, 2, 3, 4],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]})

df2 = pd.DataFrame({'Year':[1, 2, 4, 5],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]})

# Inner join
inner_join = pd.merge(df1, df2, on = 'Year', how='inner')
inner_join
```

```python
# Left join (keep indexes on the left - df1), putting NaN in the columns we have no values
left_join = pd.merge(df1, df2, on = 'Year', how='left')
left_join.set_index('Year', inplace=True)
left_join
```

```python
# We could fill all NaN with a given number
left_join.fillna(value=0.0)
```

```python
# Outer join (keeps all), putting NaN in the columns we have no values
outer_join = pd.merge(df1, df2, on = 'Year', how='outer')
outer_join.set_index('Year', inplace=True)
outer_join
```

```python
# If we drop all NaN rows it would be the same as inner join
outer_join.dropna()
```
