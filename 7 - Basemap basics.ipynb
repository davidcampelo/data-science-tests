## Using basemap module to plot geographic maps
(Inspired by a tutorial published at www.pythonprogramming.net)

The objective of this tutorial is to cover the basics of the matplotlib's extension basemap to plot some maps.

Before all, it's necesary to install this module. Please follow these instructions to install it on Mac OS: https://luomiyang.wordpress.com/2014/12/03/get-started-with-basemap-library-for-python/


```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Draw a basic map in high resolution (it may take some time)
# lon_0, lat_0 are the center point of the projection.
# resolution = 'l' means use low resolution coastlines.
m = Basemap(projection='ortho',lon_0=-8,lat_0=40,resolution='l')
m.drawcoastlines()
m.fillcontinents(color='coral',lake_color='aqua')
# draw parallels and meridians.
m.drawparallels(np.arange(-90.,120.,30.))
m.drawmeridians(np.arange(0.,420.,60.))
m.drawmapboundary(fill_color='aqua')
plt.title("World map projection")
plt.show()
```

```python
# plotting coordinates
fig = plt.figure(figsize=(8,16))
m = Basemap(projection='mill',
            llcrnrlat = -40,
            llcrnrlon = -80,
            urcrnrlat = +60,
            urcrnrlon = +20,
            resolution='l')

m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')
#m.drawcounties(color='darkred')
m.fillcontinents(color='coral',lake_color='aqua')
m.drawmapboundary(fill_color='aqua')

#m.etopo()
#m.bluemarble()

xs = []
ys = []
# Cities coordinates
FORlat, FORlon = -3.7319, -38.5267
OPOlat, OPOlon = 41.1579, -8.6291

# Draw city markers
xpt, ypt = m(FORlon, FORlat)
xs.append(xpt)
ys.append(ypt)
m.plot(xpt, ypt, 'c*', markersize=15)

xpt, ypt = m(OPOlon, OPOlat)
xs.append(xpt)
ys.append(ypt)
m.plot(xpt, ypt, 'g^', markersize=15)

# Drw flight route and straight line
m.plot(xs, ys, color='k', linewidth=3, label='Straight line')
m.drawgreatcircle(FORlon, FORlat, OPOlon, OPOlat, color='r', linewidth=3, label='TAP flight 36')

plt.legend(loc=4)
plt.title('Flight route from Fortaleza (FOR) to Porto (OPO)')
plt.show()
```
